```python
import cv2
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
files = os.listdir('A')
```

```python
data = []
k = 0
for i in sorted(os.listdir('.')):
    if len(i)>1:
        continue
    files = os.listdir(i)
    print(i, len(files), k)
    for f in files:
        im = cv2.imread(i+'/'+f, cv2.IMREAD_GRAYSCALE)
        im2 = cv2.resize(im, (30,36))
        im3 = im2.flatten()
        im4 = np.insert(im3, 0, k)
        data.append(im4)
    k+=1
```

```python
all_data = data
```

```python
random.shuffle(data)
```

```python
all_data+= data
```

```python
all_data = np.array(all_data)
```

```python
all_data.shape
```

```python
cols = [i for i in range(1080)]
df = pd.DataFrame(all_data, columns=['label']+cols)
```
